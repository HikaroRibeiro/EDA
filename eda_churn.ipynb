## AnÃ¡lise Bivariada

```python
# HipÃ³tese: clientes com contrato do tipo do tipo mensal sÃ£o mais propensos ao Churn
# Construir tabela de contingÃªncia entre tipo de contrato e Churn - Quantidade.
# ConclusÃ£o: Clientes que nÃ£o abandonaram o serviÃ§o e possuem contratos mensais sÃ£o 2220, e com contrato de 1 ano Ã© de 1307, e de dois anos Ã© de 1647
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Construir tabela de contingÃªncia entre tipo de contrato e Churn - ProporÃ§Ã£o.
pd.crosstab(df_churn.Churn, df_churn.Contract, normalize='index', margins=True, margins_name='Total')
```

Considerando o pÃºblico total, os clientes de contrato mensal representam 55%. PorÃ©m, se consideramos apenas os clientes que abandonaram o serviÃ§o, Churn igual a yes, 88% tinham contrato mensal. Desta forma, podemos afirmar que hÃ¡ uma correlaÃ§Ã£o entre o tipo de contrato mensal e o abandono.

```python
# Avaliar a correlaÃ§Ã£o entre duas variÃ¡veis categÃ³ricas (Qualitativas). 

# Executar um teste de hipÃ³tese chamado Chi-Square ou Qui-Quadrado de Pearson.
# Num teste de hipÃ³tese, duas hiÃ³teses sÃ£o formuladas:
# H0 (HipÃ³tese Nula): as duas variÃ¡veis sÃ£o independentes;
# H1 (HipÃ³tese Complementar): as duas variÃ¡veis nÃ£o sÃ£o independentes;

# O teste serve para confirmar ou recusar a hipÃ³tese nula (H0)
# Define um valor de controle que se for abaixo desse nÃºmero, significa que a probabilidade de encontrar a hipÃ³tese nula Ã© inferior ao nÃºmero definido.
# Recusa essa hipÃ³tes e se a hipÃ³tese complementar.
# Quando a probablilidade de observarmos H0 Ã© inferior a 0.05 (p-Value),
# recusamos a hipÃ³tese e seguimos com a complementar.
# Retorna o p-Value e um nÃ­vel de enquadrar (magnitude dessa populaÃ§Ã£o). 
# NÃºmero muito alto mas o p-Value for maior que 0.05 nÃ£o podemos recusar hipÃ³tese nula.
```

```python
# Gerar um DF da Crosstab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calcular os Scores e P-Values
# Utilizar um outro mÃ³dulo interno ao Panda. Instalar o "scipy" -> pipenv install sci
from scipy.stats import chi2_contingency
```

```python
# Tupla = Lista imutÃ¡vel
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar nÃºmeros com decimais sem a notaÃ§Ã£o cientÃ­fica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar um dataframe com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# Analisar Scores e P-Values
df_chi_scores_churn_contract
```

Conforme teste estatÃ­stico (Chi-Square), o P-Value menor <= a 0.05, desta forma rejeitamos a hipÃ³tes nula, ou seja, as variÃ¡veis nÃ£o sÃ£o independentes. Pelo Qui2 alto, podemos afirmar que hÃ¡ uma forte correlaÃ§Ã£o.

```python
# Avaliar a correlaÃ§Ã£o entre uma variÃ¡vel qualitativa e quantitativa.
# Usar o mesmo teste de Chi-Square, mas transformar a variÃ¡vel quantitativa em qualitativa.
# HipÃ³tese: Cliente com menos de 6 meses de contrato Ã© mais propenso ao Churn. (Testar essa hipÃ³tese).

# Criar uma coluna nova no Dataframe com base em uma condiÃ§Ã£o em outra coluna.
# Coluna = 'TempoMenor6Meses'.

# NecessÃ¡rio importar a biblioteca 'numpy' 

import numpy as np
```

```python
# Criar coluna nova.
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure < 6, 'Yes', 'No')
```

```python
# Checar a coluna criada:
df_churn.head(5)
```

```python
# Gerar um DF da Crosstab (sem totais).
df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
# Teste de hipÃ³tese. Calcular os scores.
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
# Criar as duas sÃ©ries:
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
# Criar um dataframe com os resultados:
df_chi_scores_churn_tenure =pd.DataFrame({'Qui2': scores_churn_tenure, 'p-value': pvalues_churn_tenure})
```

```python
# Analisar os resultados:
df_chi_scores_churn_tenure
```

Como o P-Value <=0.05, rejeitamos a H0, ou seja, as variÃ¡veis nÃ£o sÃ£o independentes. E vale mencionar que com base no Qui2 resultante das 2 anaÄºises, a correlaÃ§Ã£o entre Churn e Tempo de Contrato Ã© menor do que 6 meses Ã© menor do que a correlaÃ§Ã£o entre Churn e Tipo de Contrato.

```python
# CorrelaÃ§Ã£o entre 2 variÃ¡veis numÃ©ricas:
# Tenure = Tempo de contrato, Total_Charges = Total pago pelo cliente enquanto eles esteve com o contrato;
# Analisa o Tenure com TotalCharges;
# IntuiÃ§Ã£o, quanto maior o tempo de contrato, maior o valor pago;
# Coeficiente de correlaÃ§Ã£o de Pearson. 

# CorrelaÃ§Ã£o entre 2 variÃ¡veis numÃ©ricas - Pearson. Sem colocar argumento ele utiliza o mÃ©todo de Pearson.
df_churn.tenure.corr(df_churn.TotalCharges)
```

```python
# CorrelaÃ§Ã£o entre 2 variaveis numÃ©ricas - Spearman
df_churn.tenure.corr(df_churn.TotalCharges, method='spearman')
```
